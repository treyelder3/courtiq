```python
def find_likely_duplicates(players, dob_tolerance_days=365):
    """
    Identifies pairs of players that are likely the same person
    based on name match and similar date of birth.

    Two players are flagged as likely the same person if:
    - They share the same first and last name AND
    - Their DOBs are within dob_tolerance_days OR share the same month and day (different year)

    Returns a dataframe of likely duplicate pairs.
    """
    from itertools import combinations
    import pandas as pd

    # Get all players with duplicate names
    name_counts = players.groupby(['name_first', 'name_last'])['player_id'].transform('count')
    dup_players = players[name_counts > 1].copy()

    likely_same = []

    # Group by name
    for (first, last), group in dup_players.groupby(['name_first', 'name_last']):
        if len(group) < 2:
            continue

        # Compare all pairs within same name group
        for (idx1, row1), (idx2, row2) in combinations(group.iterrows(), 2):
            try:
                dob1 = pd.to_datetime(row1['dob'])
                dob2 = pd.to_datetime(row2['dob'])
                dob_diff = abs((dob1 - dob2).days)

                # Check if within tolerance
                within_tolerance = dob_diff <= dob_tolerance_days

                # Check if same month and day regardless of year
                same_month_day = (dob1.month == dob2.month) and (dob1.day == dob2.day)

                likely_same_person = within_tolerance or same_month_day

            except:
                # If dob is missing/unparseable, flag as uncertain
                likely_same_person = None
                dob_diff = None
                same_month_day = None

            likely_same.append({
                'player_id_1': row1['player_id'],
                'player_id_2': row2['player_id'],
                'name': f"{first} {last}",
                'dob_1': row1['dob'],
                'dob_2': row2['dob'],
                'dob_diff_days': dob_diff,
                'same_month_day': same_month_day,
                'likely_same_person': likely_same_person
            })

    return pd.DataFrame(likely_same)

# Run it
duplicate_pairs = find_likely_duplicates(players)
print(f"Total duplicate name pairs: {len(duplicate_pairs)}")
print(f"Likely same person: {duplicate_pairs['likely_same_person'].sum()}")
print(f"Likely different people: {(duplicate_pairs['likely_same_person'] == False).sum()}")
print(f"Uncertain (missing DOB): {duplicate_pairs['likely_same_person'].isna().sum()}")
duplicate_pairs.head(10)
```

```python
# Manual overrides: tournament display names that don't match city or simple normalization
MANUAL = {
    'French Open (Roland Garros)':            'Roland Garros',
    'Italian Open':                            'Rome',
    'Canadian Open':                           'Montreal',    # alternates Montreal/Toronto year-by-year
    'Monte-Carlo Masters':                     'Monte-Carlo Masters',
    'Swiss Indoors':                           'Basel',
    'Swiss Open':                              'Gstaad',
    'Swedish Open':                            'Bastad',
    'Grand Prix Hassan II':                    'Casablanca',
    'Open Occitanie':                          'Montpellier',
    'Grand Prix Auvergne-Rh\u00f4ne-Alpes':    'Lyon',
    'Lib\u00e9ma Open':                        'Rosmalen',
    'Generali Open Kitzb\u00fchel':            'Kitzbuhel',
    'Paris Masters':                           'Paris Masters',
    "Queen's Club Championships":              "Queen's Club",
    'Hall of Fame Open (1976\u20132024)':       'Newport',
    'Hellenic Championship':                   'Athens',
    "U.S. Men's Clay Court Championships":     'Houston',
    'Winston-Salem Open':                      'Winston-Salem',
    'Eastbourne International':                'Eastbourne',
    'European Open':                           'Antwerp',
    'Bavarian International':                  'Munich',
    'Bavarian International (1899\u20132024)':  'Munich',
    'SA Tennis Open (2009\u20132011)':          'Johannesburg',
    'Srpska Open (2023)':                      'Banja Luka',
    'Bett1Hulks Cologne Championship (2020)':  'Cologne',
    'Bett1Hulks Cologne Indoors (2020)':       'Cologne',
    'Moselle Open (2003\u20132025)':            'Metz',
    'Washington Open':                         'Washington',
    'Miami Open':                              'Miami Masters',
}

def strip_year(name):
    """Remove parenthetical year ranges like (2021) or (2019-2024)."""
    return re.sub(r'\s*\([^)]*\)', '', str(name)).strip()

def normalize(s):
    """Uppercase, strip common generic words, remove non-alphanumeric."""
    s = strip_year(s)
    s = re.sub(r'\b(open|international|championships|series|indoors|indoor)\b', '', s, flags=re.I)
    s = re.sub(r"[^a-z0-9\s']", ' ', s.lower())
    return re.sub(r'\s+', ' ', s).strip().upper()

print('Manual overrides and normalization helpers ready.')
```

```python
def find_tourney_name(row):
    tournament = row['tournament']
    city       = str(row['city'])

    # 1. Manual override
    if tournament in MANUAL:
        candidate = MANUAL[tournament].upper()
        if candidate in match_names_upper:
            return match_names_upper[candidate]

    # 2. Exact city match (most ATP events are keyed by city in matches_cleaned)
    for c in [p.strip() for p in city.upper().split('/')]:
        if c in match_names_upper:
            return match_names_upper[c]

    # 3. Exact normalised tournament name
    norm_t = normalize(tournament)
    if norm_t in match_names_upper:
        return match_names_upper[norm_t]

    keys = list(match_names_upper.keys())

    # 4. Fuzzy match on normalised tournament name (high cutoff)
    fuzzy_t = get_close_matches(norm_t, keys, n=1, cutoff=0.75)
    if fuzzy_t:
        return match_names_upper[fuzzy_t[0]]

    # 5. Fuzzy match on city name
    fuzzy_c = get_close_matches(city.upper(), keys, n=1, cutoff=0.82)
    if fuzzy_c:
        return match_names_upper[fuzzy_c[0]]

    return None


atp_tourney['tourney_name'] = atp_tourney.apply(find_tourney_name, axis=1)

n_matched   = int(atp_tourney['tourney_name'].notna().sum())
n_unmatched = int(atp_tourney['tourney_name'].isna().sum())
print('Matched:  ', n_matched, '/', len(atp_tourney))
print('Unmatched:', n_unmatched)
```

```python
# Step 3: Manual resolution for known conflicts
# Keep the row with the correct/current venue per tournament
overrides = {
    'BELGRADE':       'SRPC MILAN GALE MUŠKATIROVIĆ',  # current arena
    'DALLAS':         'FORD CENTER AT THE STAR',         # correct city is Dallas
    'HAMBURG':        'AM ROTHENBAUM',                   # keep M-level (Masters) entry
    'LYON':           'LDLC ARENA',                      # current venue
    'ST. PETERSBURG': 'SIBUR ARENA',                     # keep either, drop city spelling dupe
    'STOCKHOLM':      'KUNGLIGA TENNISHALLEN',            # current venue
    'STUTTGART':      'TENNIS CLUB WEISSENHOF',          # current venue
}

def resolve_dupes(df):
    deduped = []
    for name, group in df.groupby('tourney_name'):
        if len(group) == 1:
            deduped.append(group)
        elif name in overrides:
            preferred = group[group['venue'] == overrides[name]]
            deduped.append(preferred if not preferred.empty else group.iloc[[-1]])
        else:
            deduped.append(group.iloc[[-1]])  # fallback: keep last
    return pd.concat(deduped).reset_index(drop=True)

tournaments_df = resolve_dupes(tournaments_df)

# Verify
print(tournaments_df[tournaments_df.duplicated(subset=['tourney_name'], keep=False)])
print(len(tournaments_df))
```
